```python
from sklearn.datasets import fetch_openml

mnist = fetch_openml("mnist_784", version = 1, cache=True, as_frame=False)  #pandasのデータフレームにならないように設定
X, y = mnist["data"], mnist["target"]
```

```python
from sklearn.manifold import TSNE

tsne = TSNE(n_components=2, learning_rate='auto', init='random', perplexity=30.0)

X_embedded = tsne.fit_transform(X)
```

```python
print(type(X_embedded))
print(len(X_embedded))
```

```python
X_embedded_sampled = X_embedded[:10000]
y_sampled = y[:10000]
```

```python
import matplotlib.pyplot as plt
import numpy as np
plt.figure(figsize=(13,10))
plt.scatter(X_embedded_sampled[:, 0], X_embedded_sampled[:, 1], c=y_sampled.astype(np.int32), cmap="jet")
plt.axis('off')
plt.colorbar()
plt.show()
```

```python
from sklearn.decomposition import PCA

pca = PCA(n_components=2)
X_reduced = pca.fit_transform(X)
X_reduced_sampled = X_reduced[:10000]
y_sampled = y[:10000] #t=SNEと同じ
```

```python
import matplotlib.pyplot as plt
import numpy as np
plt.figure(figsize=(13,10))
plt.scatter(X_reduced_sampled[:, 0], X_reduced_sampled[:, 1], c=y_sampled.astype(np.int32), cmap="jet")
plt.axis('off')
plt.colorbar()
plt.show()
```
